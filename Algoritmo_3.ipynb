## Modifique el Algoritmo 03 y reduzca el tiempo de ejecución ≈1000x

```python
def iteracion(n):
    if n == 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y
       
    
import time
start_t = time.time()
[iteracion(x) for x in range(5_000)]
end_t = time.time()
end_t - start_t
```

```python
def fibonacci_2(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_2(n-1, memo) + fibonacci_2(n-2, memo)
    return memo[n]

import time
start_t = time.time()
[fibonacci_2(x) for x in range(5_000)]
end_t = time.time()
end_t - start_t
```

Se puede evidenciar que en el segundo metodo, el tiempo de ejecucion es mucho menor que en el primero.
